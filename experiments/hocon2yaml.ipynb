```python
import yaml
import glob

from ptls.util import get_conf, get_cls
from pyhocon import ConfigFactory, HOCONConverter
from omegaconf import OmegaConf
```

```python
for hocon_config_name in glob.glob('*/*/*.hocon'):
    name_wo_extension = hocon_config_name[:-6]
    conf_factory = ConfigFactory.parse_file(hocon_config_name, resolve=False)
    yaml_fname = name_wo_extension + '.yaml'
    with open(yaml_fname, 'w') as h:
        print(HOCONConverter.to_yaml(conf_factory), file=h)
```
